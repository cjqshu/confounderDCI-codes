```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# 确保 alpha_dep 和 alpha_ci 都是数值类型
dci_results_df_out['alpha_dep'] = pd.to_numeric(dci_results_df_out['alpha_dep'])
dci_results_df_out['alpha_ci'] = pd.to_numeric(dci_results_df_out['alpha_ci'])

# 获取唯一排序后的 alpha_dep 和 alpha_ci
alpha_dep_unique = np.sort(dci_results_df_out['alpha_dep'].unique())
alpha_ci_unique = np.sort(dci_results_df_out['alpha_ci'].unique())

# 创建网格
X_grid, Y_grid = np.meshgrid(alpha_dep_unique, alpha_ci_unique, indexing='ij')
Z_grid = dci_results_df_out.pivot(index='alpha_dep', columns='alpha_ci', values='sqrt_PEHE').values

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# 绘制三维曲面
surf = ax.plot_surface(X_grid, Y_grid, Z_grid, cmap='coolwarm', edgecolor='none', alpha=0.85)
fig.colorbar(surf, shrink=0.5, aspect=10)

# 美化等高线：设置等高线数量（levels）、颜色、线型、标签
z_offset = Z_grid.min()-0.1*(Z_grid.max()-Z_grid.min())
levels = np.linspace(Z_grid.min(), Z_grid.max(), 12)
contour = ax.contour(
    X_grid, Y_grid, Z_grid, 
    zdir='z', 
    offset=z_offset, 
    levels=levels,
    cmap='coolwarm',
    linewidths=2,
    linestyles='solid', 
    alpha=0.85
)
# 添加等高线标签
ax.clabel(contour, fontsize=10, fmt="%.2f", colors='k', inline=True)

# 叠加一层白色透明底座增强分层感，仅设置color和alpha，避免facecolors
ax.plot_surface(
    X_grid, Y_grid, np.full_like(Z_grid, z_offset), 
    color='w',      # 用color而不是facecolors
    alpha=0.05, 
    shade=False
)


# 通过 labelpad 参数增加标签与坐标轴的距离，使 label 不挨着刻度
# # ax.set_title("sqrt_PEHE 3D Surface (alpha_dep vs alpha_ci)")
ax.set_xlabel("$\\alpha_0$", labelpad=12)  # labelpad 全称是 label padding，即“标签与坐标轴的距离填充”。
ax.set_ylabel(f"$\\alpha_{0}'$", labelpad=12)
ax.set_zlabel("$\sqrt{{\epsilon_{{PEHE}}}}$", rotation=90, labelpad=12)  # 适当增大 labelpad 数值

# 手动设置主刻度间隔
from matplotlib.ticker import MultipleLocator

# 设置 x 轴和 y 轴主刻度间隔，可以根据实际数据范围进行调整
ax.xaxis.set_major_locator(MultipleLocator(0.5))
ax.yaxis.set_major_locator(MultipleLocator(0.5))
ax.zaxis.set_major_locator(MultipleLocator(0.004))

plt.tight_layout()

plt.savefig(f"./visualization/surface/DR_thresholds_PEHE.pdf", dpi=600) 
plt.show()
```
