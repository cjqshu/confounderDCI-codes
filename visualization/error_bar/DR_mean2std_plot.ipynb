```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.rcParams.update({ 
    "text.usetex": True,
    "font.family": "Times New Roman",
    "font.size": 16
})

from matplotlib.ticker import MultipleLocator, AutoMinorLocator
```

```python
csv_directory = "./visualization/error_bar/"

DR_nostd_df = pd.read_csv(csv_directory + "DR_omega_nostandard_IHDP.csv")
DR_std_df = pd.read_csv(csv_directory + "DR_omega_standard_IHDP.csv")
```

```python
# DR_nostd_df
```

```python
plt.figure(figsize=(10,6))
plt.errorbar(DR_nostd_df['OMEGA_CONF'], 
             DR_nostd_df['sqrt_PEHE'], 
             yerr=DR_nostd_df['sqrt_PEHE_CIhalf'], 
             fmt='-o', capsize=4, color='#A8363A', label='No Standardization')
plt.errorbar(DR_std_df['OMEGA_CONF'], 
             DR_std_df['sqrt_PEHE'], 
             yerr=DR_std_df['sqrt_PEHE_CIhalf'], 
             fmt='-o', capsize=4, color='#2369BC', label='Standardization')
plt.xlabel("$\omega$")
plt.ylabel("$\sqrt{\epsilon_{PEHE}}$")
# plt.title("sqrt_PEHE vs OMEGA_CONF")
plt.legend()
# plt.grid(True)

# 隐去上边和右边的刻度
plt.tick_params(axis='x', which='both', top=False)
plt.tick_params(axis='y', which='both', right=False)

# 隐去次刻度
# plt.tick_params(axis='x', which='minor', bottom=False, top=False)
plt.tick_params(axis='y', which='minor', left=False, right=False)

# 设置主刻度间隔为4, 设置次刻度间隔为1
plt.gca().xaxis.set_major_locator(MultipleLocator(0.2))
plt.gca().xaxis.set_minor_locator(MultipleLocator(0.05))

plt.annotate(r'\textbf{(a)}', xy=(0, 0), xycoords='axes fraction',
    xytext=(-0.05, 1), textcoords='axes fraction',
    fontsize=22, ha='right', va='top')
# plt.title(r"BNN-NCC", fontsize=12)


plt.savefig(f'./visualization/error_bar/DR_diff_omega_PEHE.pdf', dpi=600, bbox_inches='tight')

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(DR_nostd_df['OMEGA_CONF'], 
         DR_nostd_df['|ATE_err|'], 
         '-o', color='#A8363A', label='No Standardization')
plt.plot(DR_std_df['OMEGA_CONF'], 
         DR_std_df['|ATE_err|'], 
         '-o', color='#2369BC', label='Standardization')
plt.xlabel("$\omega$")
plt.ylabel("$\epsilon_{ATE}$")
# plt.title("|ATE_err| vs OMEGA_CONF")
plt.legend()
# plt.grid(True)

# plt.tick_params(axis='both', which='both', length=0)  # 去除刻度，但保留刻度标签

# 隐去上边和右边的刻度
plt.tick_params(axis='x', which='both', top=False)
plt.tick_params(axis='y', which='both', right=False)

# 隐去次刻度
# plt.tick_params(axis='x', which='minor', bottom=False, top=False)
plt.tick_params(axis='y', which='minor', left=False, right=False)

# 设置主刻度间隔为4, 设置次刻度间隔为1
plt.gca().xaxis.set_major_locator(MultipleLocator(0.2))
plt.gca().xaxis.set_minor_locator(MultipleLocator(0.05))

plt.annotate(r'\textbf{(b)}', xy=(0, 0), xycoords='axes fraction',
    xytext=(-0.05, 1), textcoords='axes fraction',
    fontsize=22, ha='right', va='top')
# plt.title(r"BNN-NCC", fontsize=12)

plt.savefig(f'./visualization/error_bar/DR_diff_omega_ATE.pdf', dpi=600, bbox_inches='tight')

plt.show()
```
