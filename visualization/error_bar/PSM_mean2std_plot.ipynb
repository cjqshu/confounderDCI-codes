```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from utils.plots import set_plot_style
set_plot_style(science_plots=False)
# set_plot_style(science_plots=True)

plt.rcParams.update({ 
    "text.usetex": True,
    "font.family": "Times New Roman",
    "font.size": 16
})


# 设置主刻度和次刻度
from matplotlib.ticker import MultipleLocator, AutoMinorLocator
```

```python
csv_directory = "./visualization/error_bar/"

psm_nostd_df = pd.read_csv(csv_directory + "PSM_match_k_nostandard_IHDP.csv")
psm_std_df = pd.read_csv(csv_directory + "PSM_match_k_standard_IHDP.csv")
```

```python
plt.figure(figsize=(10,6))
plt.errorbar(psm_nostd_df['MATCH_K'], 
             psm_nostd_df['sqrt_PEHE'], 
             yerr=psm_nostd_df['sqrt_PEHE_CIhalf'], 
             fmt='-o', capsize=4, color='#A8363A', label='No Standardization')
plt.errorbar(psm_std_df['MATCH_K'], 
             psm_std_df['sqrt_PEHE'], 
             yerr=psm_std_df['sqrt_PEHE_CIhalf'], 
             fmt='-o', capsize=4, color='#2369BC', label='Standardization')
plt.xlabel("$K$")
plt.ylabel("$\sqrt{\epsilon_{PEHE}}$")
# plt.title("sqrt_PEHE vs MATCH_K")
plt.legend()
# plt.grid(True)

plt.xticks(ticks=range(int(min(psm_nostd_df['MATCH_K'].min(), psm_std_df['MATCH_K'].min())),
                      int(max(psm_nostd_df['MATCH_K'].max(), psm_std_df['MATCH_K'].max())) + 1, 4))

# 隐去上边和右边的刻度
plt.tick_params(axis='x', which='both', top=False)
plt.tick_params(axis='y', which='both', right=False)

# 隐去次刻度
# plt.tick_params(axis='x', which='minor', bottom=False, top=False)
plt.tick_params(axis='y', which='minor', left=False, right=False)

# 设置主刻度间隔为4, 设置次刻度间隔为1
plt.gca().xaxis.set_major_locator(MultipleLocator(4))
plt.gca().xaxis.set_minor_locator(MultipleLocator(1))

plt.annotate(r'\textbf{(c)}', xy=(0, 0), xycoords='axes fraction',
    xytext=(-0.05, 1), textcoords='axes fraction',
    fontsize=22, ha='right', va='top')
# plt.title(r"BNN-NCC", fontsize=12)

plt.savefig(f'./visualization/error_bar/PSM_diff_K_PEHE.pdf', dpi=600, bbox_inches='tight')

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(psm_nostd_df['MATCH_K'], 
         psm_nostd_df['|ATE_err|'], 
         '-o', color='#A8363A', label='No Standardization')
plt.plot(psm_std_df['MATCH_K'], 
         psm_std_df['|ATE_err|'], 
         '-o', color='#2369BC', label='Standardization')
plt.xlabel("$K$")
plt.ylabel("$\epsilon_{ATE}$")
# plt.title("|ATE_err| vs MATCH_K")
plt.legend()
# plt.grid(True)


plt.xticks(ticks=range(int(min(psm_nostd_df['MATCH_K'].min(), psm_std_df['MATCH_K'].min())),
                      int(max(psm_nostd_df['MATCH_K'].max(), psm_std_df['MATCH_K'].max())) + 1, 4))

# 隐去上边和右边的刻度
plt.tick_params(axis='x', which='both', top=False)
plt.tick_params(axis='y', which='both', right=False)

# plt.tick_params(axis='x', which='minor', bottom=False, top=False)
plt.tick_params(axis='y', which='minor', left=False, right=False)


# 设置x轴主刻度间隔为4, 设置次刻度间隔为1
plt.gca().xaxis.set_major_locator(MultipleLocator(4))
plt.gca().xaxis.set_minor_locator(MultipleLocator(1))

# 可以对y轴同理设置（如有需要）
# plt.gca().yaxis.set_major_locator(MultipleLocator(0.2))
# plt.gca().yaxis.set_minor_locator(MultipleLocator(0.05))

plt.annotate(r'\textbf{(d)}', xy=(0, 0), xycoords='axes fraction',
    xytext=(-0.05, 1), textcoords='axes fraction',
    fontsize=22, ha='right', va='top')
# plt.title(r"BNN-NCC", fontsize=12)

plt.savefig(f'./visualization/error_bar/PSM_diff_K_ATE.pdf', dpi=600, bbox_inches='tight')

plt.show()
```
