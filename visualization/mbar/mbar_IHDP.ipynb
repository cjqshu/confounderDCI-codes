```python
# -----------------------------------------
# 数据接口1 
# -----------------------------------------
# # 原始数据，未求均值和标准差
# df_train = pd.read_csv('./visualization/mbar/ood_train_r2.csv')
# df_test = pd.read_csv('./visualization/mbar/ood_test_r2.csv')
# means_train = df_train.mean()
# stds_train = df_train.std()

# -----------------------------------------
# 数据接口2
# -----------------------------------------

# DLasso C0: 16 [16, 3, 5, 11, 18, 21]
# Dmrmr C0: 1 [1, 4, 6, 10, 20, 21]
# BackCD C0: 23 [23, 2, 11],
# Random C0: 21 [21, 1, 2, 3, 11, 16]
# Random2 C0: 20 [20, 1, 2], 
# DCorr C0: 9 [9, 3]

diff_C0_methods = ["DLasso", "DMrmr", "BackCD", "Random_1", "Random_2", "Ours"]

means_train = pd.DataFrame([1.505, 1.513, 1.506, 1.526, 1.511, 1.502], index=diff_C0_methods)
stds_train = pd.DataFrame([0.040, 0.040, 0.050, 0.050, 0.040, 0.054], index=diff_C0_methods)

# 按照 mean_values_train 从大到小排序
sorted_indices_train = means_train.iloc[:,0].sort_values(ascending=False).index  # means_train是DataFrame
means_sorted_train = means_train.loc[sorted_indices_train]
stds_sorted_train = stds_train.loc[sorted_indices_train]
methods_train = means_sorted_train.index.tolist()

mean_values_train = means_sorted_train.values.flatten()  
std_values_train = stds_sorted_train.values.flatten()


# ========= 绘制训练集柱状图 =========
fig1 = plt.figure(figsize=(10, 6))
# hatch_patterns_train = ['/', '/', '/', '/', '/', '/', '/', '/']
hatch_patterns_train = ['\\', '\\', '\\', '\\', '\\', '\\', '\\', '\\']


bar_width = 0.45
x_train = np.arange(len(methods_train))

bars1 = plt.bar(
    x_train, 
    mean_values_train, 
    yerr=std_values_train, 
    capsize=7,
    alpha=0.7, edgecolor='black', linewidth=1.4, color='#A8363A', width=bar_width
)

for i, bar in enumerate(bars1):
    bar.set_hatch(hatch_patterns_train[i % len(hatch_patterns_train)])

plt.xticks(x_train, methods_train, rotation=45)
# plt.xlabel('Method', fontsize=16, fontweight='bold')
plt.ylabel('$\sqrt{\epsilon_{PEHE}}$', fontsize=16, fontweight='bold')
# -------- 更改标题 verticalalignment --------
# plt.title('The IHDP training set $R^2$ score comparison (Environment $X_6=0$)', fontsize=22, fontweight='bold', pad=16)
plt.grid(axis='y', alpha=0.3, linestyle='--')
plt.tick_params(axis='x', labelsize=16, length=8)  # 保持x轴
plt.tick_params(axis='y', labelsize=16, length=10)  # 加长y轴刻度

# plt.text(-0.12+0.06, 1.08, '(a)', transform=plt.gca().transAxes, fontsize=24, fontweight='bold', va='top', ha='left')

# 扩大y轴范围
y_max_train = max(mean_values_train + std_values_train) + 0.1
y_min_train = min(1.2, min(mean_values_train - std_values_train) - 0.05)
plt.ylim(y_min_train, y_max_train)

for i, (mean_val, std_val) in enumerate(zip(mean_values_train, std_values_train)):
    plt.text(i, mean_val + std_val + 0.01,
            f'{mean_val:.3f}±{std_val:.3f}',
            ha='center', va='bottom', fontsize=16)

plt.tight_layout()

plt.savefig(f"./visualization/mbar/DR_PEHE_diff_C0.pdf", dpi=600) 

plt.show()
```
