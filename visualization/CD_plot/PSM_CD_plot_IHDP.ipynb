```python
import pandas as pd
```

```python
df = pd.read_csv("./visualization/CD_plot/PSM_data_CD_IHDP.csv", index_col=False)
df
```

```python
avg_rank = df.groupby('estimator')['PEHE'].mean().sort_values(ascending=False)
avg_rank
```

```python
import scikit_posthocs as sp

test_results = sp.posthoc_conover_friedman(
    df,
    melted=True,
    block_col='cv_fold',
    block_id_col='cv_fold',
    group_col='estimator',
    y_col='PEHE',
)
sp.sign_plot(test_results)
```

```python

import matplotlib.pyplot as plt
import scienceplots

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Times New Roman",
    "font.size": 12
})

# plt.figure(figsize=(10, 2), dpi=600)
plt.figure(figsize=(6, 4), dpi=600)

sp.critical_difference_diagram(
    avg_rank, test_results, 
    label_fmt_left="{label} ({rank:.3g})", 
    label_fmt_right="({rank:.3g}) {label}", 
)

plt.savefig(f'./visualization/CD_plot/PSM_CD_IHDP.pdf', dpi=600, bbox_inches='tight')

plt.show()
```
